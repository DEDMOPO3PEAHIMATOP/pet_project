## Пет-проект "Парсинг вакансий в hh.ru"

### План проекта

#### Цели, ТЗ, декомпозиция

*В рамках данного проекта планируется решить проблему поиска наиболее подходящих вакансий для трудоустройства по определенным и выверенным критериям.*

**Цель проекта:**

Разработка программы, которая находит по заданным критериям вакансии на сайте hh.ru  

### Техническое задание:
**Требования, которые должны быть выполнены для достижения результата и конечных целей:**
- Работа по написанию программы-парсера должна быть выполнена с помошью языка python;
- План работы должен состоять их трех ключевых этапов: поиск данных, извлечение информации, сохранение данных;
- Работа должна включать использование не менее трех популярных фреймворков;
- Результат работы должен быть представлен в виде файла test.csv — с результатами поиска вакансий;
- Результат работы необходимо визуализировать (EDA+дашборд).

**Срок реализации проекта:** 
*Три недели, с 12.09.2022 до 03.10.2022*
